# PCA - Principle Component Analysis

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df.head()
```

```python
df = df.drop(columns= ['User ID', 'Gender'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(columns=['Purchased']), df['Purchased'], test_size=0.2, random_state=42)
```

```python
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA(n_components=2)),
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42))
])
```

```python
pipe
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

#**PCA on Covid_toy dataset**

```python
df = pd.read_csv('covid_toy.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
si = SimpleImputer()
```

```python
df['fever'] = si.fit_transform(df[['fever']])
```

```python
lb = LabelEncoder()
df['cough'] = lb.fit_transform(df['cough'])
df['city'] = lb.fit_transform(df['city'])
df['gender'] = lb.fit_transform(df['gender'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(columns=['has_covid']), df['has_covid'], test_size=0.2, random_state=42)
```

```python
pipe = Pipeline([

    ('scaler', StandardScaler()),
    ('pca', PCA(n_components=2)),
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42))
])
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```
