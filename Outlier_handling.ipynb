```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/newplacementdata.csv')
```

```python
df.head(2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['placement_exam_marks'].describe()
```

```python
sns.boxplot(x =  df['placement_exam_marks'])
```

```python
# Finding the IRQ

percentile25 = df['placement_exam_marks'].quantile(0.25)
percentile75 = df['placement_exam_marks'].quantile(0.75)
```

```python
percentile25
```

```python
percentile75
```

```python
IQR = percentile75 - percentile25
IQR
```

```python
upper_limit = percentile75 + 1.5*IQR
upper_limit
```

```python
lower_limit = percentile25 - 1.5*IQR
lower_limit
```

#Finding our outliers

```python
df[df['placement_exam_marks']> upper_limit]
```

```python
df[df['placement_exam_marks']< lower_limit]
```

```python
# Trimming  ---> 4(lower).....1 , upper(25).....(50) ---> here 1,50 are outliers --->
# lower(4) <----> upper(25) mid data .

# Capping ---->  4(lower).....1 , upper(25).....(50) ---> capping lower point lowest point(1)
# upper point upper point(50)  ---> Range expand ---> outliers remove .
```

```python
newdf= df[df['placement_exam_marks'] < upper_limit]
```

```python
newdf
```

```python
plt.figure(figsize = (15,10))


plt.subplot(222)
sns.boxplot(x = df['placement_exam_marks'])


plt.subplot(224)
sns.boxplot(x = newdf['placement_exam_marks'])
plt.show()
```

```python
# Cappiing (Outlier removing technique 2)
```

```python
new_df_cap = df.copy()
```

```python
# min = 5  , max = 15

# min 4, 3, 1
# max = 20, 30, 50

# updated_min_value = 1
# updated_max_value = 50
```

```python
new_df_cap['placement_exam_marks'] = np.where(

    new_df_cap['placement_exam_marks'] > upper_limit ,
    upper_limit ,

    np.where(
    new_df_cap['placement_exam_marks'] < lower_limit ,
    lower_limit ,
    new_df_cap['placement_exam_marks'])


)
```

```python
new_df_cap
```

```python
plt.figure(figsize = (15,8))


plt.subplot(222)
sns.boxplot(x=df['placement_exam_marks'])


plt.subplot(224)
sns.boxplot(x=new_df_cap['placement_exam_marks'])
plt.show()
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df['Age'].describe()
```

```python
sns.boxplot(x = df['Age'])
```

```python
percentile25 = df['Age'].quantile(0.25)
percentile75 = df['Age'].quantile(0.75)
```

```python
print(percentile25)
print(percentile75)
```

```python
IQR = percentile75 - percentile25
IQR
```

```python
upper_limit = percentile75 + 1.5*IQR
upper_limit
```

```python
lower_limit = percentile25 - 1.5*IQR
lower_limit
```

```python
# finding outlier
```

```python
df[df['Age'] > upper_limit]
```

```python
df[df['Age'] < lower_limit]
```

```python
# trimming
```

```python
newdf = df[df['Age'] < upper_limit]
```

```python
plt.figure(figsize = (15,10))


plt.subplot(222)
sns.boxplot(x = df['Age'])


plt.subplot(224)
sns.boxplot(x = newdf['Age'])
```

```python
# capping
df_cap = df.copy()
```

```python
df_cap['Age'] = np.where(

    df_cap['Age'] > upper_limit ,
    upper_limit ,

    np.where(
    df_cap['Age'] < lower_limit ,
    lower_limit,
    df_cap['Age'])
)
```

```python
plt.figure(figsize = (15,10))


plt.subplot(222)
sns.boxplot(x = df['Age'])


plt.subplot(224)
sns.boxplot(x = df_cap['Age'])
```
